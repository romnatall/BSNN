```python
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    0.01,
    decay_steps=200000,
    decay_rate=0.98,
    staircase=True)
```

```python
lrelu=keras.layers.ELU()
iner=tf.keras.initializers.HeNormal()
inp=keras.Input(shape=(64,))
x=Dense(64, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(inp)
x=Dense(40, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(30, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(25, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(20, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(10, kernel_initializer=iner,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(8, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(6, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(5, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(4, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(3, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(2, kernel_initializer=iner ,kernel_regularizer=keras.regularizers.l2(0.005),activation=lrelu)(x)
x=Dense(1, kernel_initializer=iner ,activation=lrelu)(x)
model=keras.Model(inp,x)
model.compile(optimizer=keras.optimizers.SGD(learning_rate=lr_schedule),loss=tf.keras.losses.Huber(),metrics=['mean_absolute_error'])
#print(model.summary())
```
